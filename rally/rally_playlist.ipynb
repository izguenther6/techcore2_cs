```python
import re

# Read the JSON file
with open('songs_data.json', 'r') as f:
    json_data = f.read()

# Read the HTML file
with open('rally_playlist.html', 'r') as f:
    html_content = f.read()

# Find and replace the songs array in the HTML
pattern = r'const songs = \[.*?\];'
replacement = f'const songs = {json_data};'
updated_html = re.sub(pattern, replacement, html_content, flags=re.DOTALL)

# Write the updated HTML back
with open('rally_playlist.html', 'w') as f:
    f.write(updated_html)

print("✅ HTML playlist updated with latest songs!")
```
